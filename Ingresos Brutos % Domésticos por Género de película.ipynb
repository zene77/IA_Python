```python
# Define el estilo
sns.set(style="darkgrid", palette="muted", color_codes=True)
ax = sns.boxplot(data=peliculas3, x='Genero', y='Ingresos % USA', orient='v', color='lightgray', showfliers=False)
plt.setp(ax.artists, alpha=0.5)
```

```python
# Agrega los puntos para mostrar cada observaciÃ³n
sns.stripplot(x='Genero', y='Ingresos % USA', data=peliculas3, jitter=True, size=6, linewidth=0, hue = 'Estudio', alpha=0.7)
```

```python
# Define el estilo
sns.set(style="darkgrid", palette="muted", color_codes=True)
ax = sns.boxplot(data=peliculas3, x='Genero', y='Ingresos % USA', orient='v', color='lightgray', showfliers=False)
plt.setp(ax.artists, alpha=0.5)
# Agrega los puntos para mostrar cada observaciÃ³n
sns.stripplot(x='Genero', y='Ingresos % USA', data=peliculas3, jitter=True, size=6, linewidth=0, hue = 'Estudio', alpha=0.7)
ax.axes.set_title('Ingresos Brutos (%) Domesticos',fontsize=30)
ax.set_xlabel('Genero',fontsize=20)
ax.set_ylabel('Ingresos % Usa',fontsize=20)
ax.legend(bbox_to_anchor=(1.05, 1), loc=2)
```
