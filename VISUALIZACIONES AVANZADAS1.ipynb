CONSTRUIENDO DESHBOARDS BONUS

```python

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
# Configurar el estilo
sns.set_style("darkgrid")

# Crear subplots
f, axes = plt.subplots(2, 2, figsize=(15, 15))

# Primer grÃ¡fico KDE
k1 = sns.kdeplot(x='PresupuestoMillones', y='RatingsAudience', data=peliculas, shade=True, ax=axes[0, 0], cmap='Blues')
axes[0, 0].set_title('Presupuesto y Ratings de Audiencia')

# Segundo grÃ¡fico KDE
k2 = sns.kdeplot(x='PresupuestoMillones', y='RatingsAudience', data=peliculas, shade=True, ax=axes[0, 1], cmap='Blues')
axes[0, 1].set_title('Presupuesto y Ratings de CrÃ­ticos')

# Tercer grÃ¡fico Violin Plot
sns.violinplot(data=peliculas[peliculas.Genero == 'Drama'], x='AÃ±o', y='RatingsCriticos', ax=axes[1, 0])
axes[1, 0].set_title('Ratings de CrÃ­ticos por AÃ±o en Drama')

# Cuarto grÃ¡fico KDE
k4 = sns.kdeplot(x='RatingsCriticos', y='RatingsAudience', data=peliculas, shade=True, shade_lowest=False, cmap='Greens', ax=axes[1, 1])
k4b = sns.kdeplot(x='RatingsCriticos', y='RatingsAudience', data=peliculas, cmap='Greens', ax=axes[1, 1])
axes[1, 1].set_title('Ratings de CrÃ­ticos y Audiencia')

# Establecer los lÃ­mites x
axes[0, 0].set_xlim(-20, 160)
axes[0, 1].set_xlim(-20, 160)

# Mostrar los grÃ¡ficos
plt.show()
```

```python
# Configurar el estilo
sns.set_style("darkgrid")

# Crear subplots
f, axes = plt.subplots(2, 2, figsize=(15, 15))

# Primer grÃ¡fico KDE
k1 = sns.kdeplot(x='PresupuestoMillones', y='RatingsAudience', data=peliculas, shade=True, shade_lowest=True, cmap='inferno', ax=axes[0, 0])
axes[0, 0].set_title('Presupuesto y Ratings de Audiencia')

# Segundo grÃ¡fico KDE
k2 = sns.kdeplot(x='PresupuestoMillones', y='RatingsCriticos', data=peliculas, shade=True, shade_lowest=True, cmap='inferno', ax=axes[0, 1])
axes[0, 1].set_title('Presupuesto y Ratings de CrÃ­ticos')

# Tercer grÃ¡fico Violin Plot con paleta de colores 'inferno'
sns.violinplot(data=peliculas[peliculas.Genero == 'Drama'], x='AÃ±o', y='RatingsCriticos', ax=axes[1, 0], palette='inferno')
axes[1, 0].set_title('Ratings de CrÃ­ticos por AÃ±o en Drama')

# Cuarto grÃ¡fico KDE con fondo color inferno
k4 = sns.kdeplot(x='RatingsCriticos', y='RatingsAudience', data=peliculas, shade=True, shade_lowest=False, cmap='inferno', ax=axes[1, 1])
k4b = sns.kdeplot(x='RatingsCriticos', y='RatingsAudience', data=peliculas, cmap='cool', ax=axes[1, 1])
axes[1, 1].set_title('Ratings de CrÃ­ticos y Audiencia')

# Establecer los lÃ­mites x
axes[0, 0].set_xlim(-35, 260)
axes[0, 1].set_xlim(-35, 260)

# Mostrar los grÃ¡ficos
plt.show()
```

```python
# Configurar el estilo
sns.set_style("dark",{"axes.facecolor":"black"})

# Crear subplots
f, axes = plt.subplots(2, 2, figsize=(15, 15))

# Primer grÃ¡fico KDE
k1 = sns.kdeplot(x='PresupuestoMillones', y='RatingsAudience', data=peliculas, shade=True, shade_lowest=True, cmap='inferno', ax=axes[0, 0])
axes[0, 0].set_title('Presupuesto y Ratings de Audiencia')

# Segundo grÃ¡fico KDE
k2 = sns.kdeplot(x='PresupuestoMillones', y='RatingsCriticos', data=peliculas, shade=True, shade_lowest=True, cmap='inferno', ax=axes[0, 1])
axes[0, 1].set_title('Presupuesto y Ratings de CrÃ­ticos')

# Tercer grÃ¡fico Violin Plot con paleta de colores 'inferno'
sns.violinplot(data=peliculas[peliculas.Genero == 'Drama'], x='AÃ±o', y='RatingsCriticos', palette='inferno', ax=axes[1, 0])
axes[1, 0].set_title('Ratings de CrÃ­ticos por AÃ±o en Drama')

# Cuarto grÃ¡fico KDE con fondo correcto
k4 = sns.kdeplot(x='RatingsCriticos', y='RatingsAudience', data=peliculas, shade=True, shade_lowest=True, cmap='inferno', ax=axes[1, 1])
k4b = sns.kdeplot(x='RatingsCriticos', y='RatingsAudience', data=peliculas, cmap='cool', ax=axes[1, 1])
axes[1, 1].set_title('Ratings de CrÃ­ticos y Audiencia')

# Establecer los lÃ­mites x
axes[0, 0].set_xlim(-35, 260)
axes[0, 1].set_xlim(-35, 260)

# Mostrar los grÃ¡ficos
plt.show()
```
