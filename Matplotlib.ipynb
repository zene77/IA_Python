```python
import matplotlib.pyplot as plt

# Datos de ejemplo
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

# Primer subplot
plt.subplot(1, 2, 1)
plt.plot(x, y, 'g')  # Grfico en verde
plt.title('Grfico 1')

# Segundo subplot
plt.subplot(1, 2, 2)
plt.plot(y, x, 'blue')  # Grfico en azul
plt.title('Grfico 2')

# Mostrar los grficos
plt.tight_layout()
plt.show()
```

## Multigraficos (filas, columnas)

```python
figura, graficos = plt.subplots(nrows=4,ncols=2)
```

```python
figura, graficos = plt.subplots(nrows=2,ncols=2)

graficos[0][0].plot(x,y, 'blue' )
graficos[0][1].plot(x,y, 'red' )
graficos[1][0].plot(x,y, 'y' )
graficos[1][1].plot(x,y, 'g' )

graficos[0][0].set_title('Azul' )
graficos[0][1].set_title('Rojo' )
graficos[1][0].set_title('Amarillo' )
graficos[1][1].set_title('Verde' )
```

## Dos grficos en la misma figura

```python
figura = plt.figure()
graficos = figura.add_axes([0.1, 0.1, 0.9, 0.9])
graficos.plot(x, x**2, label='Al cuadrado')
graficos.plot(x, x**3, label= 'Al cubo')
# Aadir leyenda
graficos.legend()

# Mostrar los grficos
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos de ejemplo
x = np.linspace(-10, 10, 400)

# Crear la figura
figura = plt.figure()

# Crear los ejes
graficos = figura.add_axes([0.1, 0.1, 0.8, 0.8])

# Aadir grficos
graficos.plot(x, x**2, label='Al cuadrado')
graficos.plot(x, x**3, label='Al cubo')

# Aadir ttulo y etiquetas
graficos.set_title('Grficos de $x^2$ y $x^3$')
graficos.set_xlabel('x')
graficos.set_ylabel('y')

# Aadir leyenda
graficos.legend()

# Mostrar los grficos
plt.show()
```

## Color de grafico

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos de ejemplo
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Crear el grfico
plt.plot(x, y, 'red')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos de ejemplo
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Crear el grfico con una lnea roja
plt.plot(x, y, color='red')
plt.title('Grfico de Seno con Lnea Roja')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```

## Linea del grafico (grosor, transferencia y tipo)

```python
figura = plt.figure()
graficos = figura.add_axes([0, 0, 1, 1])
graficos.plot(x, y, 'red', linewidth=5)
plt.show() 
```

```python
figura = plt.figure()
graficos = figura.add_axes([0, 0, 1, 1])
graficos.plot(x, y, 'red', linewidth=5, alpha= 0.5)
plt.show() 
```

```python
figura = plt.figure()
graficos = figura.add_axes([0, 0, 1, 1])
graficos.plot(x, y, 'red', linewidth=5, alpha= 0.5, linestyle='-.')
plt.show() 
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos de ejemplo
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Crear el grfico
plt.plot(x, y, color='green', marker='o', markersize=5)
plt.title('Grfico con Marcadores')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid(True)
plt.show()
```
