## Graficos de cuadrícula: PairGird y FacetGrid

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
flores = sns.load_dataset('iris')
```

```python
flores.head()
```

```python
sns.pairplot(flores)
```

```python
graficos = sns.PairGrid(flores)
graficos.map(plt.scatter)
```

```python
graficos = sns.PairGrid(flores)
graficos.map_diag(sns.histplot)
graficos.map_upper(plt.scatter)
```

```python
# Suponiendo que 'flores' es un DataFrame con datos
# Crear el PairGrid
graficos = sns.PairGrid(flores)

# Mapear histogramas a la diagonal
graficos.map_diag(sns.histplot)
graficos.map_lower(sns.kdeplot)

# Mostrar el gráfico
plt.show()
```

```python
graficos = sns.PairGrid(flores)
graficos.map_diag(sns.histplot)
graficos.map_upper(plt.scatter)
graficos.map_lower(sns.kdeplot)
```

## FcetGrid

```python
propinas = sns.load_dataset('tips')
propinas.head()
```

```python
propinas['sex'].unique()
```

```python
propinas['time'].unique()
```

```python
graficos2 = sns.FacetGrid(data=propinas, col='time', row='smoker')
graficos2.map(sns.histplot, 'total_bill',kde=True)
```

```python
graficos2 = sns.FacetGrid(data=propinas, col='time', row='smoker')
graficos2.map(sns.scatterplot, 'total_bill', 'tip' )
```

```python
# Suponiendo que 'propinas' es un DataFrame con datos
# Crear el FacetGrid
graficos2 = sns.FacetGrid(data=propinas, col='time', row='smoker')

# Mapear gráficos de dispersión
graficos2.map(sns.scatterplot, 'total_bill', 'tip')

# Mostrar el gráfico
plt.show()
```
